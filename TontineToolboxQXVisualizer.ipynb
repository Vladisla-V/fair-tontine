```python
from example_mortality_data import MALE_QX, FEMALE_QX
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
QX Table + Gompertz mortality visualizer. Linear + Log scales.

m = Modal (peak) age of deaths. If M == 90 --> the most common age people die in is 90.
b = Dispersion parameter. How "spread out" the deaths are around that modal age

Milevsky & Salisbury (2015/2016) Gompertz parameters:
m = 88.72
b = 10.0
"""

def plot_mortality_analysis(M, b):
    # Setup ages
    ages = np.arange(len(MALE_QX))
    
    # Calculate Gompertz theoretical values
    # mu(x) = (1/b) * exp((x - m) / b)
    theoretical_mu = (1/b) * np.exp((ages - m) / b)

    plt.style.use('dark_background')
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # --- Plotting Linear Scale ---
    ax1.plot(ages, MALE_QX, color='steelblue', label='Male Qx (Empirical)')
    ax1.plot(ages, FEMALE_QX, color='firebrick', label='Female Qx (Empirical)')
    ax1.plot(ages, theoretical_mu, label='Gompertz Model (Paper)', color='white', linestyle='--')
    
    ax1.set_title(f'Mortality Rate: Linear Scale (m={m}, b={b})')
    ax1.set_xlabel('Age')
    ax1.set_ylabel('Probability of Death / Force of Mortality')
    ax1.legend()
    ax1.grid(True, alpha=0.15)

    # --- Plotting Logarithmic Scale ---
    ax2.semilogy(ages, MALE_QX, color='steelblue', label='Male Qx (Empirical)')
    ax2.semilogy(ages, FEMALE_QX, color='firebrick', label='Female Qx (Empirical)')
    ax2.semilogy(ages, theoretical_mu, label='Gompertz Model (Paper)', color='white', linestyle='--')
    
    ax2.set_title(f'Mortality Rate: Log Scale (m={m}, b={b})')
    ax2.set_xlabel('Age')
    ax2.set_ylabel('log(Mortality)')
    ax2.legend()
    ax2.grid(True, which="both", alpha=0.15)

    plt.tight_layout()
    plt.show()

# Change parameters here
m = 88.72
b = 10.0

plot_mortality_analysis(m, b)
```
